## Q4 (10 points)
You are given with an unsorted array ARR. Your task is to sort the array in wave format.

Explanation:

1. If ARR have N elements then, sorted ARR should be like:-

   ARR[i] >= ARR[i+1] and ARR[i+1] <= ARR[i+2]

  Example:   
   ARR[0] >= ARR[1] and ARR[1] <= ARR[2],

   ARR[2] >= ARR[3] and ARR[3] <= ARR[4],

   ARR[4] >= ARR[5] and ARR[5] <= ARR[6],  and so on...


Input Format:

1. First line of the input takes number 'T' denoting the number if test cases.

2. Next 2*T lines represent T test cases.

3. First line of each test case takes integer input 'N' i.e. number of elements an array contain.

4. Second line of each test case takes array input of N spaced elements.


Output Format:

A single line contains array sorted in wave format.  



Example:

INPUT: 
```
2
7
5 3 3 2 1 2 4 
6
60 51 11 9 50 57
```
OUTPUT:


```
4 3 5 2 3 1 2
50 11 60 51 57 9
```

Note: 
1. There can we more than one pattern of sorted wave array, any right pattern is acceptable.
2. Try to do as much optimised code as possible.

````python
# Write your code here
def waveArray(array,size):
    array.sort()
    size = int((size/2)*2)
    
    for i in range(0,size-1,2):
        temp =array[i]
        array[i]=array[i+1]
        array[i+1]=temp





t=0
t = int(input("Number of testcases"))
while(t>0):
    size = int(input("Size of array"))
    array=[]
    for i in range(0,size):
        array.append(int(input('Enter numbers')))
    waveArray(array,size)
    for i in range(0,len(array)):
        print(array[i])
    t-=1
 
````
